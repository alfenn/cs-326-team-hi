```python
# We only care about ConvertedCompYearly

display(df[['ConvertedCompYearly', 'CompTotal']])

# - ConvertedCompYearly: float64
# - Currency: object (142 unique values)
# - CompTotal: float64
# - Country: object (185 unique values)

# Drop all null values in 'ConvertedCompYearly' column in df
# 65437 -> 23435 rows
df = df.dropna(subset=['ConvertedCompYearly'])

display(df)
```

```python
import pprint
# Drop rows where respondants aren't full-time employed professional engineers

# - Employment: object (110 unique values)
unique_vals = df['Employment'].unique()
val_set = set()
for v in unique_vals:
    vals = v.split(';')
    for v2 in vals:
        val_set.add(v2)
# pprint.pprint(df['Employment'].unique())
pprint.pprint(val_set)

# We filter by these values
# - 'Employed, full-time'
# - 'Independent contractor, freelancer, or self-employed'

def is_full_time(s):
    return 'Employed, full-time' in s or 'Independent contractor, freelancer, or self-employed' in s

# Apply the is_full_time function to each row in the 'Employment' column
# If the function returns False, drop the row, otherwise keep it
df = df[df['Employment'].apply(is_full_time)]

# 23435 -> 22436 rows
display(df)
```
