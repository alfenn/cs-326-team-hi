```python
sorted_value_counts = []
import pprint

def exclude_col(s):
    exclude_cols = ['SOHow', 'LogConvertedCompYearly', 'ConvertedCompYearly', 'YearsCodePro', 'YearsCode']
    for colname in exclude_cols:
        if colname in s:
            return True
    return False

for column in train_df.columns:
    if exclude_col(column): 
        continue
    value_counts = train_df[column].value_counts().sort_index()
    for value, count in value_counts.items():
        sorted_value_counts.append([count, f'{column}: {value}'])
        
sorted_value_counts.sort()
for cnt, col_val in sorted_value_counts:
    print(f"{col_val} - {cnt}")
```
